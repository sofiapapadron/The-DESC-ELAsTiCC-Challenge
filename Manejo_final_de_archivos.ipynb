```python
# Importar las librerías necesarias
import os
import tarfile
import urllib.request

# Definir la URL y el nombre de archivo
url = 'https://portal.nersc.gov/cfs/lsst/DESC_TD_PUBLIC/ELASTICC/TRAINING_SAMPLES/FULL_ELASTICC_TRAIN.tar'
filename = 'FULL_ELASTICC_TRAIN.tar'

# Descargar el archivo
urllib.request.urlretrieve(url, filename)

# Crear una carpeta para extraer los archivos si no existe
if not os.path.exists('extracted_files'):
    os.makedirs('extracted_files')

# Extraer el archivo .tar
with tarfile.open(filename, 'r') as tar:
    tar.extractall(path='extracted_files')

# Listar el contenido de la carpeta extraída
extracted_files = os.listdir('extracted_files')
print('Archivos y carpetas extraídas:', extracted_files)
```

Concatenación de archivos

```python
import os
import gzip
import shutil

# Ruta a la carpeta donde están los archivos
folder = 'extracted_files/ELASTICC_TRAIN_EB'

# Archivos de salida
head_output_file = 'EB_HEAD.FITS.gz'
phot_output_file = 'EB_PHOT.FITS.gz'

# Función para concatenar archivos
def concatenate_files(folder, output_file, pattern):
    with gzip.open(output_file, 'wb') as f_out:
        for file_name in sorted(os.listdir(folder)):
            if pattern in file_name:  # Verificar si el archivo es del tipo correcto
                file_path = os.path.join(folder, file_name)
                with gzip.open(file_path, 'rb') as f_in:
                    shutil.copyfileobj(f_in, f_out)
                print(f"Concatenando: {file_name}")

# Concatenar archivos HEAD
concatenate_files(folder, head_output_file, 'HEAD')

# Concatenar archivos PHOT
concatenate_files(folder, phot_output_file, 'PHOT')

print("Concatenación completada.")
```

```python
from astropy.io import fits
import csv

# Función para convertir archivos FITS a CSV
def fits_to_csv(input_fits_file, output_csv_file):
    # Abrir el archivo FITS
    with fits.open(input_fits_file) as hdul:
        data = hdul[1].data  # Asumimos que los datos están en la segunda extensión

        # Abrir el archivo CSV para escribir los datos
        with open(output_csv_file, 'w', newline='') as csvfile:
            writer = csv.writer(csvfile)

            # Escribir los nombres de las columnas
            writer.writerow(data.names)

            # Escribir los datos fila por fila
            for row in data:
                writer.writerow(row)

# Archivos de entrada (descomprime primero si es necesario)
head_fits_file = 'EB_HEAD.FITS.gz'
phot_fits_file = 'EB_PHOT.FITS.gz'

# Archivos de salida en formato CSV
head_csv_file = 'EB_HEAD.csv'
phot_csv_file = 'EB_PHOT.csv'

# Convertir FITS a CSV
fits_to_csv(head_fits_file, head_csv_file)
fits_to_csv(phot_fits_file, phot_csv_file)

print("Conversión completada. Archivos CSV generados.")
```
